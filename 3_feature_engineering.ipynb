```python
import pandas as pd 
import numpy as np

pd.__version__
```

```python
INPUT_DATA = 'data/4_final_dataset/hourly_weather_events_data.csv'
OUTPUT_DATA = 'data/4_final_dataset/final_dataset.csv'
```

```python
df_merged = pd.read_csv(INPUT_DATA, delimiter=';')
df_merged.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_merged[10000:10060]
```

### `is_alarm`
- a boolean value showcasing whether there is an alarm going off in the given region at given time

```python
df_merged['is_alarm'] = df_merged['event_start_hour'].apply(lambda x: x == x)
```

```python
df_merged.head(10)
```

### `alarms_countrywide`
- the number of regions where the alarm is on at the moment

```python
df_merged_v2 = df_merged.copy()
regions_w_alarm_on = df_merged_v2[df_merged_v2['is_alarm']==True].groupby(['day_datetime', 'hour_datetime'])['region'].count()
```

```python
regions_feature = pd.DataFrame(regions_w_alarm_on)
regions_feature.head(15)
```

```python
regions_feature = regions_feature.rename(columns={'region': 'alarms_countrywide'})
df_merged_v2 = df_merged_v2.merge(regions_feature, how='left', on=['day_datetime', 'hour_datetime'], )
df_merged_v2.shape
```

```python
# manually checking the correctness of the feature
df_merged_v2[df_merged_v2['day_datetime']=='2022-02-26'][df_merged_v2['hour_datetime']=='17:00:00']
```

```python
#replacing NaN's with 0's
df_merged_v2['alarms_countrywide'] = df_merged_v2['alarms_countrywide'].apply(lambda x: int(x) if x == x else 0)
```

```python
df_merged_v2['alarms_countrywide'].describe()
```

### `n_alarms_last24hrs`
- the number of alarms in the given region during the last 24 hrs

```python
df_merged_v3 = df_merged_v2.copy()
# df_merged_v3.dtypes
df_merged_v3.shape
```

```python
df_merged_v3['datetime'] = pd.to_datetime(
    df_merged_v3['day_datetime'].str.cat(
        df_merged_v3['hour_datetime'], 
        sep=' '))
```

```python
df_merged_v3.head(5)
```

```python
df_alarms = df_merged_v3[['region', 'datetime', 'event_start_time']]
df_alarms['dup'] = df_alarms.duplicated(subset=['region', 'event_start_time'], keep='first')
df_alarms.head(100)
```

```python
df_alarms['dup'] = df_alarms.apply(
    lambda row: not pd.isna(row['event_start_time']) if pd.isna(row['event_start_time']) else row['dup'], axis=1)
```

```python
df_alarms.head(100)
```

```python
df_alarms['event_start_time'] = df_alarms.apply(
    lambda row: np.nan if row['dup'] else row['event_start_time'], axis=1)
```

```python
df_alarms.head(100)
```

```python
df_alarms_v1 = df_alarms.groupby('region')[['datetime', 'event_start_time']].rolling('24H', on='datetime').count()
df_alarms_v1 = pd.DataFrame(df_alarms_v1).reset_index()
df_alarms_v1 = df_alarms_v1.drop('level_1', axis=1)
df_alarms_v1 = df_alarms_v1.rename(columns={'event_start_time':'n_alarms_last24hrs'})
df_alarms_v1.head(200)
```

```python
df_alarms_v1.shape
```

```python
df_merged_v4 = pd.merge(df_merged_v3,
    df_alarms_v1, 
    on=['region', 'datetime'])
df_merged_v4.tail(300)
```

```python
df_merged_v4['n_alarms_last24hrs'].isna().sum()
```

```python
df_merged_v4.shape
```

```python
df_merged_v4 = df_merged_v4.drop_duplicates()
```

```python
df_merged_v4['n_alarms_last24hrs'] = df_merged_v4['n_alarms_last24hrs'].apply(lambda x: int(x))
```

```python
df_merged_v4.tail(100)
```

### `is_weekend`
- a boolean value showing whether the given day is a weekend

```python
df_merged_v5 = df_merged_v4.copy()
df_merged_v5.dtypes
```

```python
df_merged_v5['day_datetime'] = pd.to_datetime(df_merged_v5['day_datetime'])
```

```python
df_merged_v5['is_weekend'] = df_merged_v5.apply(lambda row: row['day_datetime'].day_name() in ['Saturday', 'Sunday'], axis=1)
```

```python
df_merged_v5.head(100)
```

### `region_of_Ukraine`
- categorizes oblasts by region (east / west / north / south / central)

```python
# IDs for different regions of Ukraine
ID_north = [6, 10, 18, 25]
ID_south = [1, 21, 8, 14, 4]
ID_west = [2, 3, 7, 9, 13, 17, 19, 22, 24]
ID_east = [1, 4, 5, 12, 14, 15, 20, 21]
ID_central = [2, 16, 10, 11, 23]

# New feature named region_of_Ukraine to indicate the part of the country
df_merged_v6 = df_merged_v5.assign(region_of_Ukraine='')
df_merged_v6.loc[df_merged_v6['region_id'].isin(ID_north), 'region_of_Ukraine'] = 'north'
df_merged_v6.loc[df_merged_v6['region_id'].isin(ID_south), 'region_of_Ukraine'] = 'south'
df_merged_v6.loc[df_merged_v6['region_id'].isin(ID_west), 'region_of_Ukraine'] = 'west'
df_merged_v6.loc[df_merged_v5['region_id'].isin(ID_east), 'region_of_Ukraine'] = 'east'
df_merged_v6.loc[df_merged_v5['region_id'].isin(ID_central), 'region_of_Ukraine'] = 'central'
df_merged_v6['region_of_Ukraine'] = pd.Categorical(df_merged_v6['region_of_Ukraine'])                                                        


df_merged_v6.tail(100)                                                                                               
```

```python
df_merged_v6.to_csv(OUTPUT_DATA, index=False, sep=';')
```
